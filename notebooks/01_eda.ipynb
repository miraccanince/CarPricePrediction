# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = pd.read_csv('../data/raw/CarPrice_Assignment.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.hist(df['price'], bins=20)
plt.title('Price Distribution')
plt.show()
```

```python
sns.boxplot(x=df['price'])
plt.title('Price Boxplot')
plt.show()
```

```python
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.show()
price_corr = corr['price'].sort_values(ascending=False)
print(price_corr)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
numerics = df.select_dtypes(include=[np.number]).drop('price', axis=1)
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(numerics.values, i) for i in range(numerics.shape[1])]
vif["features"] = numerics.columns
print(vif)
```
